# Missing Data 

```python
feature_na = [feat for feat in df.columns if df[feat].isnull().sum() > 1]

for feat in feature_na:
    print(f'{feat} -> {np.round(df[feat].isnull().mean(),4)} % of missing values')
```

```python
for feat in feature_na:
    data = df.copy()
    
    # let make variable that indicate 1 if the observation was missing or 0 if not
    data[feat] = np.where(data[feat].isnull(),1,0)
    
    data.groupby(feat)['price'].median().plot.bar()
    plt.title(feat)
    plt.show()
```

# Numeric feature

```python
numeric_feat = [feat for feat in df.columns if df[feat].dtype != 'O'] 
df[numeric_feat].head()
```

```python
# numeric variable are of two types 
  # 1. Continous Variable 
  # 2. Discrete Variable
```

```python
# Discrete Variable

discrete_feat = [feat for feat in numeric_feat if len(df[feat].unique())<25]

print(f'length of feature -> {len(discrete_feat)}')
```

```python
df[discrete_feat].head()
```

```python
# lets find the relationship between between discrete variable and price

for feat in discrete_feat:
    data = df.copy()
    data.groupby(feat)['price'].median().plot.bar()
    plt.xlabel(feat)
    plt.ylabel('Price')
    plt.show()
  
```

```python
# Continous variable 

continous_feat = [feat for feat in numeric_feat if feat not in discrete_feat] 
print(f'length of feature -> {len(numeric_feat)}')
```

```python
for feat in continous_feat:
    sns.distplot(df[feat])
    plt.show()
```

```python
# we will be doing logarthmic transfornation

for feat in continous_feat:
    data = df.copy()
    if 0 in data[feat].unique():
        pass
    else:
        data[feat] = np.log(data[feat])
        plt.scatter(data[feat],data['price'])
        plt.xlabel(feat)
        plt.ylabel('price')
        plt.title(feat)
        plt.show()
```

# Outlier

```python
for feat in continous_feat:
    data = df.copy()
    if 0 in data[feat].unique():
        pass
    else:
        data[feat] = np.log(data[feat])
        data.boxplot(column=feat)
        plt.ylabel(feat)
        plt.title(feat)
        plt.show()
```
