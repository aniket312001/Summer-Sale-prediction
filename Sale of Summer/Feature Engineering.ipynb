```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.pandas.set_option('display.max_columns',None)
df = pd.read_csv('usefull_data.csv')
df.head()
```

```python
df.info()
```

# Missing

```python
feature_na = [feat for feat in df.columns if df[feat].isnull().sum() > 1]

for feat in feature_na:
    print(f'{feat} -> {np.round(df[feat].isnull().mean(),4)} % of missing values')
```

```python
for feat in feature_na:
    print(f'{feat} -> {df[feat].dtype}')
```

# Filling Nan

```python
 for feat in feature_na:
        if feat not in ['urgency_text','origin_country']:
            median = df[feat].median()
            df[feat] = df[feat].fillna(median)
            
            
for feat in ['urgency_text','origin_country']:
    df[feat] = df[feat].fillna('Missing')
```

```python
 df[feature_na].isnull().sum()
```

```python
df.head()
```

```python
obj_feat = [feat for feat in df.columns if df[feat].dtype == 'O']  
obj_feat
```

# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()

for feat in obj_feat:
    df[feat] = label.fit_transform(df[feat])
```

```python
df.head()
```

```python
df.drop('Unnamed: 0',inplace=True,axis=1)
```

```python
df.head()
```

# Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler

X = df.drop('price',axis=1)
y = df['price']

scale = StandardScaler()

scale_feat = scale.fit_transform(X)
```

```python
df = pd.DataFrame(scale_feat,columns=X.columns)

new_df = pd.concat([df,y],axis=1)
```

```python
new_df
```

# Feature Selection

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

X = new_df.drop('price',axis=1)
y = new_df['price']

feat_select = SelectFromModel(Lasso(alpha=0.005,random_state=0))

feat_select.fit(X,y)
```

```python
feat_select.get_support()
```

```python
feat_select = X.columns[(feat_select.get_support())]

print('Total feature -> ',X.shape[1])
print('Selected Feature -> ',len(feat_select))
# print('Feature with coefficient shrank to zero -> ',np.sum(feat_select.estimator_.coef_==0))
```

```python
new = new_df[feat_select]
```

```python
df = pd.concat([new,y],axis=1)
```
